```python
import random
import matplotlib.pyplot as plt

def generar_numero_aleatorio():
    return random.random()

def obtener_unidades_vendidas(numero_aleatorio):
    if numero_aleatorio <= 0.22:
        return 100
    elif numero_aleatorio <= 0.47:
        return 200
    elif numero_aleatorio <= 0.58:
        return 300
    elif numero_aleatorio <= 0.83:
        return 400
    elif numero_aleatorio <= 0.94:
        return 500
    else:
        return 600

def beneficio(pedido):
    numero = generar_numero_aleatorio()
    unidades_vendidas = obtener_unidades_vendidas(numero)
    U_devueltas = max(0, pedido - unidades_vendidas)
    Beneficio = unidades_vendidas * 1000 - pedido * 750 + U_devueltas * 250
    return Beneficio

def simulacion(num_simulaciones, n):
    L = []
    benef_total = 0
    for i in range(1, num_simulaciones+1):
        benef_total += beneficio(n)
        promedio = benef_total / i
        L.append(promedio)
    return L

plt.figure()

num_simulaciones = 10000

X = simulacion(num_simulaciones, 400)
plt.plot(range(1, num_simulaciones+1), X, label='Pedido: 400')

X = simulacion(num_simulaciones, 100)
plt.plot(range(1, num_simulaciones+1), X, label='Pedido: 100')

X = simulacion(num_simulaciones, 200)
plt.plot(range(1, num_simulaciones+1), X, label='Pedido: 200')

X = simulacion(num_simulaciones, 300)
plt.plot(range(1, num_simulaciones+1), X, label='Pedido: 300')

X = simulacion(num_simulaciones, 500)
plt.plot(range(1, num_simulaciones+1), X, label='Pedido: 500')

X = simulacion(num_simulaciones, 600)
plt.plot(range(1, num_simulaciones+1), X, label='Pedido: 600')

plt.xlabel('Número de simulaciones')
plt.ylabel('Beneficio promedio')
plt.title('Simulación de Beneficio Promedio')
plt.legend()
plt.show()
```
